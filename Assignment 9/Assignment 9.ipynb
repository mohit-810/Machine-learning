```python
import pandas as pd
import numpy as np
```

```python
# 1.

trainData = pd.read_csv('train.csv')
testData = pd.read_csv('test.csv')
```

```python
dataset = pd.concat([trainData, testData], ignore_index=True, sort=False)
```

```python
dataset.head()
```

```python
# 2.

dataset = dataset.dropna(subset=['Survived'])
```

```python
# 3.

dataset['Has_Cabin'] = ~dataset['Cabin'].isnull()
```

```python
dataset.head()
```

```python
# 4.

import re
title_list = list()
for x in dataset['Name']:
    x = re.search("Mr. | Mrs. | Master. | Miss.", x)
    if x is None:
        title_list.append('Special')
    else:
        title_list.append(x.group(0))
    
dataset['Title'] = title_list
```

```python
# 5.

dataset.fillna(dataset.mean(), inplace=True)
dataset = dataset.astype(object).fillna("unknown")
```

```python
dataset.head()
```

```python
bins = 7
dataset['Categorical_age'] = pd.qcut(dataset['Age'], bins)
dataset['Categorical_fare'] = pd.qcut(dataset['Fare'], bins)
```

```python
dataset.head()
```

```python
# 6.

columns = ['Cabin', 'Name', 'PassengerId', 'Ticket', 'Age', 'Fare', 'SibSp', 'Parch']
dataset.drop(columns, inplace=True, axis=1)
dataset
```

```python
# 8.

dataset = pd.get_dummies(dataset)
```

```python
# 9.

X = dataset.drop('Survived', axis=1)
y = dataset['Survived']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
```

```python
# 10.

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df.head()
```

```python
# 11.

from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
